# Explainable AI (XAI) for Transformer-based Emotion Classification

This notebook demonstrates the application of various XAI techniques on our fine-tuned RoBERTa-based emotion classification model. In this notebook, we apply:

1. **Gradient × Input** – to visualize token-level relevance.
2. **Improved Explanation with Conservative Propagation (LRP)** – to obtain better explanations by modifying the backward pass.
3. **Input Perturbation** – to evaluate the robustness of our model by systematically removing tokens and observing confidence changes.

The same set of 18 sample sentences (3 per emotion) is used for all experiments.

## Import Modules and Load Model

```python
# Import required modules and load the model, tokenizer, and device using our helper from model_loading.py.

import json
import torch
import numpy as np
import matplotlib.pyplot as plt

# Import our model loading helper
from model_loading import load_model

# Import our XAI functions from previously developed scripts
from gradient_input_xai import grad_x_input_roberta, visualize_relevance  # Basic Gradient x Input method
from lrp_conservative import lrp_explanation_roberta, visualize_lrp, visualize_lrp_heatmap  # Improved LRP methods
from perturbation import input_perturbation, visualize_perturbation  # Input perturbation code

# Load the model, tokenizer, and device
model, tokenizer, device = load_model()
print("Model, tokenizer, and device loaded successfully.")
```

## Define Sample Data

```python
# Define the list of 18 sample sentences (3 per emotion) as given.
# Each sample contains the original Russian text, an English translation for model input, and the label emotion.

sample_sentences = [
    # happiness (3)
    {"ru_text": "Всё было шикарно.", "text": "everything was gorgeous", "general_emotion": "happiness"},
    {"ru_text": "В целом, вечер у Лидии мне понравился.", "text": "in general i liked lydia evening", "general_emotion": "happiness"},
    {"ru_text": "И я искренне её полюбила.", "text": "and i sincerely fell in love with her", "general_emotion": "happiness"},
    
    # sadness (3)
    {"ru_text": "Боль, тоска в душе", "text": "pain longing in the soul", "general_emotion": "sadness"},
    {"ru_text": "Без общения больше месяца", "text": "without communication for more than a month", "general_emotion": "sadness"},
    {"ru_text": "Мазурка сербского оказалась плоским пирогом, весьма сухим.", "text": "serbian mazurka turned out to be a flat pie very dry", "general_emotion": "sadness"},
    
    # anger (3)
    {"ru_text": "У меня большие претензии к блюдам.", "text": "i have big complaints about dishes", "general_emotion": "anger"},
    {"ru_text": "И я думаю, он подверг критике то, что было блюдо со свининой.", "text": "and i think he criticized what was a dish with pork", "general_emotion": "anger"},
    {"ru_text": "Но я, если честно, руками есть не люблю.", "text": "but to be honest i don't like my hands", "general_emotion": "anger"},
    
    # surprise (3)
    {"ru_text": "Я его половую принадлежность даже не могу понять.", "text": "i can not even understand his gender", "general_emotion": "surprise"},
    {"ru_text": "что между ними возникла взаимная симпатия.", "text": "that mutual sympathy arose between them", "general_emotion": "surprise"},
    {"ru_text": "Вы знаете, я немножечко о каждом уже кое-что знаю.", "text": "you know i already know something a little about everyone", "general_emotion": "surprise"},
    
    # fear (3)
    {"ru_text": "Скажу откровенно, я не знала ведь, кто ко мне придет,", "text": "frankly i did not know who would come to me", "general_emotion": "fear"},
    {"ru_text": "Я боюсь к нему идти.", "text": "i'm afraid to go to him", "general_emotion": "fear"},
    {"ru_text": "Очень вкусно, но я боюсь, что там", "text": "very tasty but i'm afraid that there", "general_emotion": "fear"},
    
    # disgust (3)
    {"ru_text": "Желудок-то забитый у вас у всех.", "text": "the stomach is clogged with you all", "general_emotion": "disgust"},
    {"ru_text": "острое неприятие", "text": "acute rejection", "general_emotion": "disgust"},
    {"ru_text": "того блюда, которое тебе подали.", "text": "the dish that you were served", "general_emotion": "disgust"}
]

print("Sample sentences loaded. Total samples:", len(sample_sentences))
```

## XAI – Gradient × Input Analysis

```python
# Apply the basic Gradient × Input method to all sample sentences.
# For each sentence, compute the relevance scores using grad_x_input_roberta and visualize with a bar chart.

print("==== Running Gradient × Input Analysis ====")
for idx, example in enumerate(sample_sentences):
    en_text = example["text"]
    print(f"\n--- Gradient × Input: Example {idx+1} ---")
    print(f"Russian Text: {example['ru_text']}")
    print(f"English Text: {en_text}")
    
    tokens, gxi_scores, logits = grad_x_input_roberta(model, tokenizer, en_text, device=device)
    
    # Optionally compute predicted label (if model.config has id2label)
    pred_label_id = torch.argmax(torch.tensor(logits), dim=1).item()
    if hasattr(model.config, "id2label"):
        pred_label = model.config.id2label[pred_label_id]
    else:
        pred_label = str(pred_label_id)
    print(f"Predicted Emotion (Grad × Input): {pred_label}")
    
    visualize_relevance(tokens, gxi_scores, title=f"Grad × Input Relevance (Example {idx+1})")
```

## XAI – LRP (Conservative Propagation) Analysis

```python
# Apply the improved LRP explanation method (Conservative Propagation) to the same sample sentences.
print("\n==== Running LRP (Conservative Propagation) Analysis ====")
for idx, example in enumerate(sample_sentences):
    en_text = example["text"]
    print(f"\n--- LRP Explanation: Example {idx+1} ---")
    print(f"Russian Text: {example['ru_text']}")
    print(f"English Text: {en_text}")
    
    tokens, lrp_scores, logits = lrp_explanation_roberta(model, tokenizer, en_text, device=device)
    
    pred_label_id = torch.argmax(torch.tensor(logits), dim=1).item()
    if hasattr(model.config, "id2label"):
        pred_label = model.config.id2label[pred_label_id]
    else:
        pred_label = str(pred_label_id)
    print(f"Predicted Emotion (LRP): {pred_label}")
    
    visualize_lrp(tokens, lrp_scores, title=f"LRP Relevance (Example {idx+1})")
    visualize_lrp_heatmap(tokens, lrp_scores, title=f"LRP Heatmap (Example {idx+1})")
```

## XAI – Input Perturbation Analysis

```python
# Perform input perturbation on each sample sentence using the LRP relevance scores.
# This will iteratively remove (mask) tokens based on their LRP scores (least relevant first)
# and plot the model's confidence drop.
print("\n==== Running Input Perturbation Analysis ====")
from perturbation import input_perturbation, visualize_perturbation

for idx, example in enumerate(sample_sentences):
    en_text = example["text"]
    print(f"\n--- Perturbation Analysis: Example {idx+1} ---")
    print(f"Russian Text: {example['ru_text']}")
    print(f"English Text: {en_text}")
    
    # Compute LRP explanation to get relevance scores
    tokens, lrp_scores, logits = lrp_explanation_roberta(model, tokenizer, en_text, device=device)
    
    # Perform perturbation using the LRP relevance scores
    tokens_list, confidence_list = input_perturbation(model, tokenizer, en_text, lrp_scores, remove_least=True)
    
    plot_title = f"Perturbation Impact (Example {idx+1})"
    visualize_perturbation(confidence_list, title=plot_title)
```

## Summary & Save Results

```python
# Optionally, store the results from all methods for further analysis.
results = []
for idx, example in enumerate(sample_sentences):
    en_text = example["text"]
    tokens_gxi, gxi_scores, logits_gxi = grad_x_input_roberta(model, tokenizer, en_text, device=device)
    tokens_lrp, lrp_scores, logits_lrp = lrp_explanation_roberta(model, tokenizer, en_text, device=device)
    _, conf_list = input_perturbation(model, tokenizer, en_text, lrp_scores, remove_least=True)
    
    result = {
        "index": idx+1,
        "ru_text": example["ru_text"],
        "en_text": en_text,
        "gold_emotion": example["general_emotion"],
        "gradient_input": gxi_scores.tolist(),
        "lrp_scores": lrp_scores.tolist(),
        "perturbation_confidences": conf_list
    }
    results.append(result)

# with open("xai_results.json", "w", encoding="utf-8") as f:
#     json.dump(results, f, indent=2, ensure_ascii=False)

print("All XAI experiments completed. Results saved to 'xai_results.json'.")
```

In this notebook, we applied three XAI techniques to our Transformer-based emotion classification model:
1. **Gradient × Input:** We visualized token relevance using gradients.
2. **Improved LRP (Conservative Propagation):** We modified the backward pass via hooks and visualized relevance scores.
3. **Input Perturbation:** We removed tokens in order of ascending LRP relevance and plotted the confidence drop.

These methods help us understand which tokens contribute most to our model’s predictions and evaluate model robustness. The results (including bar charts and line graphs) are saved for further analysis.

Please refer to `xai_results.json` for the collected numeric results and to the figures for visual interpretation.
