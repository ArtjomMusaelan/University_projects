```python
import pandas as pd
```

ASSEMBLY AI

```python
data = pd.read_csv('../Task2 - Speech to Text/transcribed_data_assemblyAI.csv', delimiter=';')
data[0:50]
```

```python
# Add columns S, I, D
data.insert(0, 'Original', '')
data.insert(1, 'Sentence', data.pop('Sentence'))
data['S'] = 0
data['I'] = 0
data['D'] = 0

data[0:50].to_excel('assembly_WER_0-50.xlsx', index=False)
data[50:100].to_excel('assembly_WER_50-100.xlsx', index=False)
data[100:150].to_excel('assembly_WER_100-150.xlsx', index=False)
```
