```python
# importing packages
from pytubefix import YouTube
import os

# url input from youtube
yt = YouTube("https://www.youtube.com/watch?v=jcHAZLXrUaA&ab_channel=%D0%97%D0%B2%D0%B0%D0%BD%D1%8B%D0%B9%D1%83%D0%B6%D0%B8%D0%BD")

# extract only audio
video = yt.streams.filter(only_audio=True).first()

# set destination to save file
destination = ("")

# download the file
out_file = video.download(output_path=destination)
```

## 40 min audio

```python
from pydub import AudioSegment

# Load the audio file (new_file is the path of the downloaded MP3)
audio = AudioSegment.from_file(out_file, format="mp4")

# Time in minutes
time = 40

# Define the start and end times in milliseconds 3.441 start
start_time = 4.64 * 60 * 1000 # 3:00 -> 180,000 ms
end_time   = start_time + time * 60 * 1000 + 1  # 43:00 -> 2,580,000 ms

# Clip the audio segment
clipped_audio = audio[start_time:end_time]

# Export the clipped audio to a new file
clipped_file = f'{time}_minutes_clipped.mp3'
clipped_audio.export(clipped_file, format="mp3")

print("Clipped audio saved as", clipped_file)
```
