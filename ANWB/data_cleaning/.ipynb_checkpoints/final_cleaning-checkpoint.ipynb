```python


def import_df(name, columns_to_convert=[]):
    """
    Imports a CSV file into a pandas DataFrame and optionally converts specified columns to datetime format.
    
    Parameters:
    name (str): The name of the CSV file (without the .csv extension) located in the './data/' directory.
    columns_to_convert (list): A list of column names to be converted to datetime format. Default is an empty list.
    
    Returns:
    pd.DataFrame: A pandas DataFrame containing the data from the CSV file with specified columns converted to datetime.
    """
    
    # Read the CSV file into a DataFrame. The file is expected to be in the './data/' directory.
    df = pd.read_csv(f'./data/{name}.csv')
    
    # Convert specified columns to datetime format.
    for col in columns_to_convert:
        df[col] = pd.to_datetime(df[col])
    
    return df
```

```python
def import_weather_df(table_name):
    """
    Imports weather data from a CSV file, converts the 'dtg' column to datetime format, 
    sets it as the index, filters the DataFrame to include data from 2018-01-01 onwards,
    and sorts the DataFrame by the datetime index.
    
    Parameters:
    table_name (str): The name of the CSV file (without the .csv extension) containing weather data.
    
    Returns:
    pd.DataFrame: A pandas DataFrame containing weather data from 2018-01-01 onwards, with 'dtg' as the datetime index.
    """
    
    # Import the data from the CSV file using the import_df function.
    df = import_df(table_name)
    
    # Convert the 'dtg' column to datetime format.
    df['dtg'] = pd.to_datetime(df['dtg'])
    
    # Set the 'dtg' column as the index and filter the DataFrame to include data from 2018-01-01 onwards.
    df = df.set_index('dtg').loc['2018-01-01':, :]
    
    # Sort the DataFrame by the datetime index.
    df = df.sort_index()
    
    return df
```

#### Cleaning categorical data and missing values

Next step will be to clean categorical columns to eleminate uneccesary whitespace or other unexpected special characters.Moreover,  we will unifiy categorical values  casing to lower. On top of that, name columns will be also unified to reduce future technical debt of the project


```python
def clean_categorical_data(df):
    """
    Cleans categorical data in a DataFrame by stripping whitespace, removing punctuation, converting to lowercase,
    and renaming columns to lowercase with underscores instead of spaces.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame with categorical data to be cleaned.
    
    Returns:
    pd.DataFrame: A pandas DataFrame with cleaned categorical data.
    """
    
    # Identify columns with data type 'object' (i.e., string columns).
    string_cols = [col for col in df.columns if 'object' == str(df[col].dtype)]
    
    # Clean string columns by stripping whitespace, removing punctuation, and converting to lowercase.
    for col in string_cols:
        df[col] = df[col].str.strip()
        df[col] = df[col].str.replace(r'[^\w\s]', '', regex=True)
        df[col] = df[col].str.lower()
    
    # Rename columns to lowercase and replace spaces with underscores.
    for col in string_cols:
        df = df.rename(columns={col: str(col).lower().replace(' ', '_')})
    
    return df
```

```python
def delete_empty_columns(df):
    """
    Deletes columns from a DataFrame that are entirely empty (i.e., all values are NaN).
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame from which empty columns will be deleted.
    
    Returns:
    pd.DataFrame: A pandas DataFrame with empty columns removed.
    """
    
    # Initialize an empty list to hold the names of columns to drop.
    columns_to_drop = []
    
    # Iterate over each column in the DataFrame.
    for col in df.columns:
        # Check if the entire column is filled with NaN values.
        if df[col].isna().sum() == df.shape[0]:
            # Add the column name to the list of columns to drop.
            columns_to_drop.append(col)
    
    # Drop the columns that are entirely empty and return the modified DataFrame.
    return df.drop(columns=columns_to_drop)
```

```python

def check_df_missing_values(df):
    """
    Checks and displays the total number of missing values in the DataFrame.
    If there are missing values, it also displays the number of missing values per column.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame to check for missing values.
    """
    total_missing_values = df.isna().sum().sum()
    display(f'Total number of missing values: {total_missing_values}')
    
    if total_missing_values > 0:
        display('Number of missing values in particular columns: ')
        display(df.isna().sum())

```

Next step will be to see value distribution for every dataframe column to get basic understanding of available data. Moreover, function will show most frequent and least frequent value counted by the value_counts() pandas function

```python
def show_value_counts(df, col):
    """
    Displays the value counts of a specified column in the DataFrame.
    If there are more than 6 unique values, only the top 6 most frequent values are displayed.
    Also shows the least used values and creates a plot based on the column's data type.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the column.
    col (str): The column name to display value counts for.
    """
    display(f'Value counts of {col}')
    sorted_val_df = df[col].value_counts().sort_values(ascending=False)
    
    if sorted_val_df.shape[0] > 6:
        sorted_val_df = sorted_val_df.iloc[:6]
    display(sorted_val_df)

    print_line_break()
    
    display(f'Least used values in {col} column: ')
    display(df[col].value_counts().sort_values(ascending=True).iloc[:5])

    unique_vals_in_col = len(pd.unique(df[col]))
    col_dtype = str(df[col].dtype)
    
    if col_dtype.startswith('int') or col_dtype.startswith('float'):
        fig, ax = plt.subplots(figsize=(14, 8))
        sns.boxplot(x=col, data=df, ax=ax)
        plt.show()
    elif unique_vals_in_col < 20 and col_dtype.startswith('object'):
        fig, ax = plt.subplots(figsize=(18, 8))
        
        missing_vals = df[col].isna().sum()
        if missing_vals > 0:
            ax.axhline(y=missing_vals, color='r', linestyle='--', linewidth=2, label='Missing values in df')
            ax.legend()
            sns.countplot(x=col, data=df.replace({np.nan: 'unknown'}), ax=ax)
        sns.countplot(x=col, data=df, ax=ax)
        
        plt.show()
```

Now all the functions are combined into one show dataframe column value counts to present clear data distribution for every column

```python

def show_dataframe_column_value_counts(df):
    """
    Iterates over all columns in the DataFrame and displays value counts for each column.
    Also checks for and displays missing values for each column.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame to display value counts for.
    """
    cols = df.columns
    
    for col in cols:
        print_line_break()
        show_value_counts(df, col)
        missing_vals_in_col = df[col].isna().sum()
        
        if missing_vals_in_col > 0:
            display(f'Missing values in {col}')
            display(f'{col}: {missing_vals_in_col}')
```

#### Managing duplicate values

```python
def show_duplicated_values_in_column(df, col_name):
    """
    Displays information about duplicated values in a specified column of a DataFrame.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the column to check for duplicates.
    col_name (str): The name of the column to check for duplicated values.
    """
    display(f"Show duplicated values in column: {col_name}")
    
    # Calculate the total number of duplicated values in the column.
    total_duplicated_values = df[col_name].duplicated().sum()
    
    if total_duplicated_values > 0:
        display(f'Duplicated values in {col_name}:')
        display('Number of duplicated values / all rows')
        
        # Calculate the percentage of duplicated values.
        duplicated_values_perc = round((total_duplicated_values / df[col_name].shape[0] * 100), 2)
        display(f'{total_duplicated_values}/{df[col_name].shape[0]} : which is around {duplicated_values_perc}%')
        
        # Get the value counts of the column and filter to show only those values that are duplicated.
        sorted_val_df = df[col_name].value_counts().sort_values(ascending=False)
        sorted_val_df = sorted_val_df[sorted_val_df > 1]
        
        display(sorted_val_df)

        # Extract the duplicated values.
        duplicated_values = sorted_val_df.reset_index()[col_name]

        display('Show duplicated column rows:')
        # Display the rows that contain duplicated values in the specified column.
        display(df[df[col_name].isin(duplicated_values.to_list())])
    else:
        display('No duplicated values in this column!!!')
```

```python
def show_general_duplicate_values(df, col_name=None):
    """
    Displays information about duplicated values in a DataFrame.
    If a column name is provided, it shows duplicated values for that specific column.
    Otherwise, it shows duplicated values for the entire DataFrame.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame to check for duplicates.
    col_name (str, optional): The column name to check for duplicated values. Default is None.
    """
    if col_name is not None:
        show_duplicated_values_in_column(df, col_name)
    else:
        total_duplicated_values = df.duplicated().sum()
        if total_duplicated_values > 0:
            display(f'Duplicated values in df:')
            display('Number of duplicated values / all rows')
            duplicated_values_perc = round((total_duplicated_values / df.shape[0] * 100), 2)
            display(f'{total_duplicated_values}/{df.shape[0]} : which is around {duplicated_values_perc}%')
        else:
            display('No duplicated values in this dataframe !!!')

def drop_duplicates_in_df(df, columns):
    """
    Drops duplicate rows in a DataFrame based on the specified columns and displays the remaining duplicated values.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame to drop duplicates from.
    columns (list): A list of column names to consider for identifying duplicates.
    
    Returns:
    bool: True if duplicates were successfully dropped, False otherwise.
    """
    drop_duplicated = False

    if len(columns) > 1:
        for col in columns:
            display(f'Duplicated values in {col} after dropping them')
            print_line_break()
            drop_duplicated = df.drop_duplicates(subset=[col], inplace=True)
            show_duplicated_values_in_column(df, col)
            print_line_break()
    else:
        display(f'Duplicated values in {columns[0]} after dropping them')
        print_line_break()
        drop_duplicated = df.drop_duplicates(subset=[*columns], inplace=True)
        show_duplicated_values_in_column(df, columns[0])
    
    return drop_duplicated
```

```python
def plot_columns(df, columns, plot):
    """
    Plots specified columns of a DataFrame using the provided plotting function.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the data to plot.
    columns (list): A list of column names to plot.
    plot (function): A plotting function (e.g., sns.boxplot, sns.countplot) to use for the plots.
    """
    if len(columns) == 0:
        display('No columns to plot')
        print_line_break()
        print_line_break()
        return

    cols_length = len(columns)
    
    # Create a subplot for each column to plot
    fig, axes = plt.subplots(nrows=cols_length, ncols=1, figsize=(12, cols_length * 6), sharex=False, sharey=False)
    print(axes, type(axes))

    if not isinstance(axes, np.ndarray):
        axes = np.array([axes])
        
    # Plot each column using the provided plotting function
    for idx, current_ax in enumerate(axes.flatten()):
        if idx < len(columns):
            current_col = columns[idx]
            current_ax.set_title(f'Column: {current_col}')
            plot(x=df[current_col], ax=current_ax)

    plt.show()

def plot_numeric_columns(df, columns):
    """
    Plots numeric columns of a DataFrame using box plots.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the data to plot.
    columns (list): A list of numeric column names to plot.
    """
    plot_columns(df, columns, sns.boxplot)

def plot_string_columns(df, columns):
    """
    Plots string (categorical) columns of a DataFrame using count plots.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the data to plot.
    columns (list): A list of string (categorical) column names to plot.
    """
    plot_columns(df, columns, sns.countplot)

def plot_bool_columns(df, columns):
    """
    Plots boolean columns of a DataFrame using count plots.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the data to plot.
    columns (list): A list of boolean column names to plot.
    """
    plot_columns(df, columns, sns.countplot)

def plot_value_distributions_in_df(df, columns_to_avoid=[]):
    """
    Plots the value distributions of numeric, string (categorical), and boolean columns in a DataFrame,
    excluding specified columns.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the data to plot.
    columns_to_avoid (list): A list of column names to exclude from plotting.
    """
    # Identify columns to plot based on their data types and exclusion list
    numeric_cols = [col for col in df.columns if ('float' in str(df[col].dtype) or 'int' in str(df[col].dtype)) and col not in columns_to_avoid]
    string_cols = [col for col in df.columns if 'object' == str(df[col].dtype) and col not in columns_to_avoid]
    bool_cols = [col for col in df.columns if 'bool' == str(df[col].dtype) and col not in columns_to_avoid]

    # Plot numeric columns
    if numeric_cols:
        display('Numerical columns plotted:')
        plot_numeric_columns(df, numeric_cols)
        print_line_break()
        print_line_break()
        print_line_break()

    # Plot string columns
    if string_cols:
        display('String columns plotted:')
        plot_string_columns(df, string_cols)
        print_line_break()
        print_line_break()
        print_line_break()

    # Plot boolean columns
    if bool_cols:
        display('Bool columns plotted:')
        plot_bool_columns(df, bool_cols)
```

Thanks to these general overview it is easier to detect skewed or unequal value distributions

```python
def show_outliers_fraction(df, col, Q1, Q3, IQR):
    """
    Displays the fraction of outliers in a specified column of a DataFrame.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the data.
    col (str): The column name to check for outliers.
    Q1 (float): The first quartile (25th percentile) of the column.
    Q3 (float): The third quartile (75th percentile) of the column.
    IQR (float): The interquartile range (IQR) of the column.
    """
    print_line_break()
    display(f'The fraction of outliers in {col}')
    
    # Create a mask to identify outliers
    total_outliers_number_in_col_mask = (df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)
    total_outliers_number_in_col = df[total_outliers_number_in_col_mask].shape[0]
    
    if total_outliers_number_in_col <= 0:
        display(f'No outliers detected in {col} column')
        return

    # Calculate and display the fraction of outliers
    total_outliers_number_in_col_perc = round((total_outliers_number_in_col / df.shape[0]), 2) * 100
    display(f'{total_outliers_number_in_col} / {df.shape[0]} which is around {total_outliers_number_in_col_perc}%')
    print_line_break()

def delete_outliers(df, columns, multiplier=1.5):
    """
    Removes outliers from specified columns in a DataFrame using the IQR method.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the data.
    columns (list): A list of column names to check for and remove outliers.
    multiplier (float, optional): The multiplier for the IQR to define outlier thresholds. Default is 1.5.
    
    Returns:
    pd.DataFrame: A new DataFrame with outliers removed from the specified columns.
    """
    df_no_outliers = df.copy()
    
    for col in columns:
        Q1 = df_no_outliers[col].quantile(0.25)
        Q3 = df_no_outliers[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - (IQR * multiplier)
        upper_bound = Q3 + (IQR * multiplier)

        # Display the fraction of outliers for the current column
        show_outliers_fraction(df, col, Q1, Q3, IQR)
        
        # Print detailed outlier thresholds for the current column
        print(f"{col}: Q1={Q1}, Q3={Q3}, IQR={IQR}, Lower Bound={lower_bound}, Upper Bound={upper_bound}")

        # Remove outliers from the current column
        df_no_outliers = df_no_outliers[(df_no_outliers[col] >= lower_bound) & (df_no_outliers[col] <= upper_bound)]

    return df_no_outliers
```

These functions will showcase fractions of outliers on particular using IQR method with 1.5 IQR multiplier.  After that, all outliers will be dropped from the datasets

Due to low amount of outliers located in dataset they will be removed using IQR and quantiles lower 0.25 and greater than 0.75

#### Simplifying columns

Next step will be to simpliy particular columns in dataset. Therefore, convert_column_to_binary will keep usually 2 or 3 most frequent values and other value types are then transformed into "other category". This simplification technique will be very useful due to significant skeweness of value distribution in specific columns. That transformation will increase the loss of model's awareness and can affect model's ability to generalize learnings on training dataset therefore these potential issues will be addressed in later stage of the project 

```python
def convert_column_to_binary(df, columns_with_new_values):
    """
    Converts specified columns in a DataFrame to binary format based on given conditions.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame where columns are to be converted.
    columns_with_new_values (dict): A dictionary where keys are column names and values are dictionaries
                                    containing 'top_values' (number of most frequent values to consider)
                                    and 'new_value' (replacement value for non-top values).
    """
    for key, val in columns_with_new_values.items():
        col = key
        multiple_values = val['top_values']
        new_replace_value = val['new_value']
        
        # Get the most frequent values in the column
        most_frequent_values = df[col].value_counts().index[:multiple_values]
        
        # Apply a lambda function to replace values not in the most frequent list with new_replace_value
        df[col] = df[col].apply(lambda row: row if str(row) in most_frequent_values else new_replace_value)
```

#### Handling data incosistencies


Next step will be to clip all numerical values into same amount of fractional digits. 
Moroever, all numerical values will be transfomed to float data type and will be converted to non-negative values

```python
def clip_numerical_cols(df, columns):
    """
    Rounds numerical columns in a DataFrame to two decimal places.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing numerical columns to round.
    columns (list): A list of column names (strings) to round to two decimal places.
    """
    for col in columns:
        df[col] = df[col].round(2)


def clean_numerical_cols(df):
    """
    Cleans numerical columns in a DataFrame by converting negative values to positive and ensuring data type consistency.
    
    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing numerical columns to clean.
    """
    numeric_cols = [col for col in df.columns if ('float' in str(df[col].dtype) or 'int' in str(df[col].dtype))]
    
    for col in numeric_cols:
        df[col] = df[col].abs()  # Convert negative values to positive
        df[col] = df[col].astype(float)  # Ensure the column is of type float
```

```python
import pandas as pd
import numpy as np

def calculate_rolling_average(weather_df, value_column):
    """
    Calculates the rolling average of a specified column in a DataFrame with a 1-hour window.

    Parameters:
    weather_df (pd.DataFrame): The input pandas DataFrame containing weather data.
    value_column (str): The column name in `weather_df` for which to calculate the rolling average.

    Returns:
    pd.DataFrame: The input `weather_df` DataFrame with an added 'rolling_mean' column containing the rolling average.
                 The index is reset for subsequent operations.
    """
    # Calculate the rolling average with a 1-hour window, closed='both' includes both ends
    weather_df['rolling_mean'] = weather_df[value_column].rolling('1h', closed='both').mean()
    
    # Reset index for merging later
    return weather_df.reset_index()


def merge_driving_with_weather_df(driving_df_original, weather_df_original, value_column, new_value_column_name):
    """
    Merges driving event data with weather data by matching each event with the nearest weather data point within a 1-hour window.

    Parameters:
    driving_df_original (pd.DataFrame): The original pandas DataFrame containing driving event data.
    weather_df_original (pd.DataFrame): The original pandas DataFrame containing weather data.
    value_column (str): The column name in `weather_df_original` for which to calculate the rolling average.
    new_value_column_name (str): The name of the new column to store the rolling average in `driving_df_original`.

    Returns:
    pd.DataFrame: A new DataFrame with merged data from `driving_df_original` and `weather_df_original`.
                 It includes a new column `new_value_column_name` with the rolling average of `value_column`.
    """
    # Copy original DataFrames to avoid modifying them directly
    driving_df = driving_df_original.copy()
    weather_df = weather_df_original.copy()

    # Calculate the rolling average in the weather DataFrame
    weather_df = calculate_rolling_average(weather_df, value_column)
    
    # Convert 'event_start' column in driving_df to datetime if not already
    driving_df['event_start'] = pd.to_datetime(driving_df['event_start'])
    
    # Sort both DataFrames by their respective datetime columns for merge_asof
    weather_df = weather_df.sort_values('dtg')
    driving_df = driving_df.sort_values('event_start')
    
    # Use merge_asof to match each driving event with the nearest weather data point within a 1-hour window
    merged_df = pd.merge_asof(driving_df,
                              weather_df[['dtg', 'rolling_mean']],
                              left_on='event_start',
                              right_on='dtg',
                              direction='backward',
                              tolerance=pd.Timedelta(hours=1))
    
    # Rename the rolling mean column to the desired new_value_column_name
    merged_df = merged_df.rename(columns={'rolling_mean': new_value_column_name})
    
    # Drop the 'dtg' column from the weather DataFrame to avoid confusion
    merged_df = merged_df.drop(columns=['dtg'])
    
    return merged_df
```

```python
def scale_numerical_data(df, columns):
    """
    Scales specified numerical columns in a DataFrame using a manually implemented StandardScaler.

    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing numerical columns to scale.
    columns (list): A list of column names (strings) to scale.

    Returns:
    None
    """
    # Define a manual StandardScaler class
    class StandardScaler:
        def fit_transform(self, data):
            mean = data.mean()
            std = data.std()
            scaled = (data - mean) / std
            return scaled

    # Initialize the StandardScaler class
    scaler = StandardScaler()

    # Scale each column in the list 'columns'
    for col in columns:
        # Apply scaling to the column
        df[col] = scaler.fit_transform(df[[col]])
```

```python
def show_dist_for_cols(df, cols, boxplot=False):
    """
    Displays distribution plots (either boxplots or countplots) for specified columns in a DataFrame.

    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the data to be plotted.
    cols (list): A list of column names (strings) to plot.
    boxplot (bool): If True, plots boxplots; if False, plots countplots. Default is False.

    Returns:
    None
    """
    # Create subplots based on the number of columns to plot
    fig, axes = plt.subplots(nrows=len(cols), ncols=1, figsize=(20, 15))

    # Handle case when only one column is passed
    if len(cols) == 1:
        axes = np.array([axes])

    # Iterate over each column and plot the distribution
    for idx, ax in enumerate(axes.flatten()):
        if boxplot:
            # Plot boxplot
            sns.boxplot(x=cols[idx], ax=ax, data=df)
        else:
            # Plot countplot
            sns.countplot(x=cols[idx], ax=ax, data=df)

    # Display the plots
    plt.show()
```

## Now the data needs to be analyzed

We will calculate  a weighted mean of accident_severity table to make it a new column for safe_driving_df
So in our case the weighted mean of types of accidents severity will help us to assess if the street is high or low risk . Weighted mean will consist of    "injury_of_fatal" sum accidents 2017-2023 and  "material_damage_only" sum of accidents 2017-2023

```python
def transform_acc_sev_col_to_encoding(df):
    """
    Transform the 'accident_severity' column into one-hot encoding and join it back to the DataFrame.

    Parameters:
    df (pd.DataFrame): The input pandas DataFrame containing the 'accident_severity' column.

    Returns:
    pd.DataFrame: The transformed DataFrame with one-hot encoded 'accident_severity' column.
    """
    df = df.copy()  # Make a copy of the DataFrame to avoid modifying the original
    
    # Perform one-hot encoding on 'accident_severity' column and join it back to the DataFrame
    df = df.join(pd.get_dummies(df['accident_severity'], dtype=float))
    
    return df


def w_avg(row, weights):
    """
    Calculate the weighted average based on specified weights.

    Parameters:
    row (pd.Series): A pandas Series representing a row of data from the DataFrame.
    weights (list): A list of weights to use for weighted average calculation.

    Returns:
    float: Weighted average value.
    """
    w1, w2 = weights
    
    # Calculate weighted sum
    values_with_w_sum = row['injury_or_fatal_sum'] * w1 + row['material_damage_only_sum'] * w2
    
    # Calculate weighted average
    return values_with_w_sum / (w1 + w2)


def calc_weighted_mean_of_acc_severity(df):
    """
    Calculate the weighted mean of accident severity based on grouped data.

    Parameters:
    df (pd.DataFrame): The input pandas DataFrame with encoded accident severity columns.

    Returns:
    pd.DataFrame: A new DataFrame with 'street', 'injury_or_fatal_sum', 'material_damage_only_sum',
                 and 'weighted_avg' columns.
    """
    df = df.copy()  # Make a copy of the DataFrame to avoid modifying the original
    
    # Transform accident severity column to one-hot encoding
    df = transform_acc_sev_col_to_encoding(df)

    # Group by 'street' and aggregate sums for 'injury or fatal' and 'material damage only'
    new_df = df.groupby(['street']).agg(
        injury_or_fatal_sum=('injury or fatal', 'sum'),
        material_damage_only_sum=('material damage only', 'sum')
    ).reset_index()

    # Calculate weighted average using custom function w_avg
    new_df['weighted_avg'] = new_df.apply(lambda row: w_avg(row, [2, 1]), axis=1)

    # Print summary statistics of 'weighted_avg' column
    print(new_df['weighted_avg'].describe())

    return new_df
```

```python
streets_with_accidents_ratio_df = calc_weighted_mean_of_acc_severity(accidents_17_23_df)
```

```python
from pathlib import Path
import os

def export_cleaned_data_to_csv(df_list=[], df_names=[]):
    """
    Export cleaned DataFrames to CSV files in a specified directory.

    Parameters:
    df_list (list): A list of pandas DataFrames to export.
    df_names (list): A list of corresponding names for each DataFrame.

    Returns:
    None
    """
    for idx, df in enumerate(df_list):
        # Create directory './data_cleaned' if it doesn't exist
        if not Path('./data_cleaned').exists():
            os.mkdir('./data_cleaned/')
        
        # Export DataFrame to CSV file in './data_cleaned' directory
        df.to_csv(f'./data_cleaned/{df_names[idx]}.csv', index=False)

        
```
