### Exploring the Impact of Cybersecurity Training Frequency on Employee Adherence to Cybersecurity Practices in Small Hotels

## 1. Objective 
This analysis aims to investigate the impact of cybersecurity training frequency, knowledge increase after training, and perceived training relevancy on the effectiveness of knowledge implementation among employees in small and medium-sized enterprises. The goal is to identify key factors that contribute to improved cybersecurity practices within these organizations.

**1. Research Question**

Does the frequency of cybersecurity training affect employee adherence to cybersecurity practices in small hotels?

---

**2. Hypothesis**

$H_0$: The frequency of cybersecurity training does not significantly affect employee adherence to cybersecurity practices. <br>
$H_1$: The frequency of cybersecurity training significantly affects employee adherence to cybersecurity practices.

To begin the analysis, I reloaded the cleaned survey data, selected relevant columns, mapped categorical responses to numeric values, and generated summary statistics to ensure a clean and structured dataset with no missing values in key variables.

```python
# Reload the data to work with the original file
df = pd.read_csv('cleaned_survey.csv')  

# Keep only the necessary columns
df = df[['frequency_training', 'Knowledge_increase_after_training', 'implementation', 'relevancy', 'increase_in_effectiveness']].copy()

# Drop rows where all target columns are NaN to focus on available data
df.dropna(subset=['Knowledge_increase_after_training', 'implementation', 'increase_in_effectiveness'], inplace=True)

# Map categorical responses to numeric values
df['Knowledge_increase_after_training'] = df['Knowledge_increase_after_training'].map({
    'Not at all': 1, 'Slightly': 2, 'Moderately': 3, 'Significantly': 4
}).astype(float)

df['implementation'] = df['implementation'].map({
    'Rarely': 1, 'Sometimes': 2, 'Moderately': 3, 'Always': 4
}).astype(float)

df['increase_in_effectiveness'] = df['increase_in_effectiveness'].map({
    'No': 1, 'Not sure': 2, 'Yes': 3
}).astype(float)

# Check if there are any remaining NaN values after converting categories to numbers
print("\nRemaining missing values after conversion:")
print(df[['Knowledge_increase_after_training', 'implementation', 'increase_in_effectiveness']].isna().sum())

# Generate full statistics for numerical variables
print("\nSummary statistics for numerical variables:")
numerical_vars = ['Knowledge_increase_after_training', 'implementation', 'increase_in_effectiveness']
summary_numerical = df[numerical_vars].describe()
display(summary_numerical)

# Display distribution for categorical variables, including NaN
print("\nFrequency distribution for categorical variables:")
categorical_vars = ['frequency_training', 'relevancy']
for var in categorical_vars:
    print(f"\n{var} distribution:")
    display(df[var].value_counts(dropna=False))
```

#### 4.3 Training Effectiveness Analysis Using ANOVA and Kruskal-Wallis Tests

This analysis explores the impact of training frequency on various outcomes, including knowledge gain, implementation, relevancy, and effectiveness. Using both ANOVA and Kruskal-Wallis tests, the code examines statistically significant differences across training frequency groups to reveal key insights into the effectiveness of training programs.

---


The survey data included categorical responses (e.g., "Slightly," "Moderately") that needed to be converted into numeric values for analysis. I defined mappings to assign each categorical response a numeric value, which enabled statistical tests to work with these variables.

```python
import pandas as pd
import scipy.stats as stats

# Load the survey data
survey_data = pd.read_csv('cleaned_survey.csv')

# Encode categorical responses to numeric values for analysis
encoding_mappings = {
    "Knowledge_increase_after_training": {'Not at all': 0, 'Slightly': 1, 'Moderately': 2, 'Significantly': 3},
    "implementation": {'Rarely': 0, 'Sometimes': 1, 'Moderately': 2, 'Always': 3},
    "relevancy": {'Not very relevant': 0, 'Neutral': 1, 'Relevant': 2, 'Very relevant': 3},
    "increase_in_effectiveness": {'No': 0, 'Not sure': 1, 'Yes': 2},
    "frequency_training": {'Never': 0, 'Annually': 1, 'Quarterly': 2, 'Monthly': 3}
}

# Apply encoding to the columns
for col, mapping in encoding_mappings.items():
    survey_data[col] = survey_data[col].map(mapping)
```

The encoding process involved:

- Defining a dictionary of mappings (encoding_mappings) where each categorical response is mapped to a numeric value.
- Using a loop to apply these mappings across each relevant column. This way, we ensure the data is consistently encoded, which is essential for later statistical analysis.
---

To maintain a complete dataset, I filled in missing values in the columns, choosing the most suitable approach for each variable:

- Median Imputation: For `Knowledge_increase_after_training`, `implementation`, and `relevancy`, I used the median to fill missing values, preserving the distribution.
- Mode Imputation: For `frequency_training`, I used the mode (most common response), which is logical for categorical data.

```python
# Impute missing values
survey_data['Knowledge_increase_after_training'] = survey_data['Knowledge_increase_after_training'].fillna(survey_data['Knowledge_increase_after_training'].median())
survey_data['implementation'] = survey_data['implementation'].fillna(survey_data['implementation'].median())
survey_data['relevancy'] = survey_data['relevancy'].fillna(survey_data['relevancy'].median())
survey_data['frequency_training'] = survey_data['frequency_training'].fillna(survey_data['frequency_training'].mode()[0])
```

I checked for any remaining missing values to ensure the dataset was fully prepared for analysis.

```python
# Check for missing values
print("Missing values per column after encoding and imputation:")
print(survey_data[encoding_mappings.keys()].isna().sum())
```

To understand the distribution of each variable, I calculated descriptive statistics (count, mean, and standard deviation) for all five variables.

```python
print("\nDescriptive Statistics:")
for variable in encoding_mappings.keys():
    count = survey_data[variable].count()
    mean = survey_data[variable].mean()
    std = survey_data[variable].std()
    print(f"{variable} - count: {count}, mean: {mean:.2f}, std: {std:.2f}")
```

This step allowed me to view the basic distributional characteristics of each variable and check for any anomalies or patterns.

---

To compare the variables between groups, I used both ANOVA and Kruskal-Wallis tests:

- ANOVA: This test provided an F-statistic and p-value to show statistically significant differences between groups.
- Kruskal-Wallis: I used this test as a non-parametric alternative when ANOVA assumptions weren’t met.

```python
# ANOVA Results
print("\nANOVA Results:")
for variable in encoding_mappings.keys():
    values = [survey_data[survey_data['frequency_training'] == level][variable].dropna()
              for level in survey_data['frequency_training'].unique()]
    if len(values) > 1:
        f_stat, p_val = stats.f_oneway(*values)
        print(f"{variable}: F-statistic = {f_stat:.2f}, p-value = {p_val:.4f}")
    else:
        print(f"{variable}: ANOVA not possible due to insufficient data")

# Kruskal-Wallis Results
print("\nKruskal-Wallis Results:")
for variable in encoding_mappings.keys():
    values = [survey_data[survey_data['frequency_training'] == level][variable].dropna()
              for level in survey_data['frequency_training'].unique()]
    if len(values) > 1:
        h_stat, p_val = stats.kruskal(*values)
        print(f"{variable}: H-statistic = {h_stat:.2f}, p-value = {p_val:.4f}")
    else:
        print(f"{variable}: Kruskal-Wallis not possible due to insufficient data")
```
